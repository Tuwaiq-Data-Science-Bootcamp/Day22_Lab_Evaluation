```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_excel('Example_ModelEval.xlsx')
df
```

```python
df2 = df
df2 = df2.assign(Absolute_Error = np.abs(df2['Predicted Score - M1'] - df2['Exam Score']))
df2 = df2.assign(Relative_Error = np.abs(df2['Predicted Score - M1'] - df2['Exam Score'])/df2['Exam Score'])
df2 = df2.assign(Square_Error = np.square(df2['Predicted Score - M1'] - df2['Exam Score']))
df2
```

```python
import sklearn.metrics as metrics
mae = metrics.mean_absolute_error(df2['Exam Score'], df2['Predicted Score - M1'])
mse = metrics.mean_squared_error(df2['Predicted Score - M1'], df2['Exam Score'])
rmse = np.sqrt(mse)
r2 = metrics.r2_score(df2['Predicted Score - M1'], df2['Exam Score'])

print("Model1 Evaluation  :")
print("MAE:",mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-Squared:", r2)
```
